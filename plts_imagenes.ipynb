Radial Distribution Function
===

```python
import numpy as np
import matplotlib.pyplot as plt
''''
'los datos para esta aprooximacioon fueron obtenidos de la siguiente fuente:'
'https://www.researchgate.net/publication/232066145_A_study_of_the_influence_of_isotopic_substitution_on_the_melting_point_and_temperature_of_maximum_density_of_water_by_means_of_path_integral_simulations_of_rigid_models'
'''

# Datos experimentales aproximados de RDF de O-H en agua
r_values = np.linspace(0, 3.8, 100)  # Distancias en Å
#radial distribution O-H
g_r = 1.29*np.exp(-((r_values-1.875)/0.6)**2) + 1.17*np.exp(-((r_values-3.225)/0.6)**2)+ 1.1*np.exp(-((r_values-4.25)/0.5)**2)  # Forma aproximada



# Graficar RDF
# Ajustes de fuente y color de texto
plt.rcParams['font.family'] = 'Arial'
plt.rcParams['text.color'] = '#464747'
plt.figure(figsize=(8, 6), dpi=200)
plt.plot(r_values, g_r, label="Experimental O-H RDF (290K, 1 bar)")
plt.xlabel("Distance r (Å)", fontsize=12, fontweight='bold')
plt.ylabel("g(r)", fontsize=12, fontweight='bold')
plt.legend()
plt.title("Radial Distribution Function of O-H in Water", fontsize=14, fontweight='bold')
#plt.grid()
plt.show()
```
